```python
# transform Income column to a numerical
df_clean['Income'] = df_clean['Income'].str.replace('$', '')
df_clean['Income'] = df_clean['Income'].str.replace(',', '').astype('float')
```

```python
#changing the DT_Customer from text to DateTime
df_clean['Dt_Customer'] = pd.to_datetime(df_clean['Dt_Customer'])
```

```python
to_drop=["AcceptedCmp1", "AcceptedCmp2", "AcceptedCmp3", "AcceptedCmp4", "AcceptedCmp5", "Response", "Complain"]
```

```python
df_clean.drop(to_drop, inplace=True, axis=1)
```

```python
df_clean.tail(10)
```

```python
df_clean= pd.DataFrame(df_clean, columns=["ID",'Year_Birth','Education','Marital_Status','Income','Country','Kidhome',
 'Teenhome','Dt_Customer',
 'Recency',
 'MntWines',
 'MntFruits',
 'MntMeatProducts',
 'MntFishProducts',
 'MntSweetProducts',
 'MntGoldProds',
 'NumDealsPurchases',
 'NumWebPurchases',
 'NumCatalogPurchases',
 'NumStorePurchases',
 'NumWebVisitsMonth'])
```

```python
df_clean.head(10)
```

```python
df_clean['Dependents'] = df_clean['Kidhome'] + df_clean['Teenhome']
```

```python
# Year becoming a Customer
df_clean['Year_Customer'] = pd.DatetimeIndex(df_clean['Dt_Customer']).year
```

```python
# Total Amount Spent
mnt_cols = [col for col in df_clean.columns if 'Mnt' in col]
df_clean['TotalMnt'] = df_clean[mnt_cols].sum(axis=1)
```

```python
df_clean.head(10)
```

```python
df_clean = df_clean[["ID", "Country", "Year_Birth", "Education", "Income", "Marital_Status", "Dependents", 
"Dt_Customer", "Year_Customer", "TotalMnt", "MntWines", "MntFruits", "MntMeatProducts", "MntFishProducts",  
"MntSweetProducts", "MntGoldProds", 'NumDealsPurchases', "NumWebPurchases", "NumCatalogPurchases", 
"NumStorePurchases", "NumWebVisitsMonth", ]] 
```
